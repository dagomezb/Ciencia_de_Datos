# **Relplot**

Gráficos de relación entre variables numéricas

En este ejercicio, exploraremos el conjunto de datos mpg, que contiene una fila por modelo de coche e incluye información como el año de fabricación del coche, el número de millas por galón ("M.P.G.") que alcanza, la potencia de su motor (medida en "caballos") y su país de origen.

¿Cuál es la relación entre la potencia del motor de un coche ("caballos") y su eficiencia de combustible ("mpg")? ¿Y cómo varía esta relación según el número de cilindros ("cylinders") que tenga el coche?

```python
sns.relplot(data=mpg,x="horsepower",y="mpg",size="cylinders")

plt.show()
```

```python
#Utilizando el canal de color (hue) para la variable cylinders

sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders",
            hue="cylinders")

# Show plot
plt.show()
```

## Lineplot

Otro tipo de gráfico que nos permite evaluar tendencia.

¿Cuál ha sido el comportamiento de mpg a través del tiempo?

```python
sns.relplot(x="model_year", y="mpg", 
            data=mpg, kind="line")

plt.show()
```

**Pregunta**
¿Cuál de las siguientes NO es una interpretación correcta de este gráfico lineal?

*Posibles respuestas*

El promedio de millas por galón ha aumentado generalmente con el tiempo.

La distribución de millas por galón es menor en 1973 comparado con 1977.

El intervalo de confianza del 95% para el promedio de millas por galón en 1970 es aproximadamente 16 - 19.5 millas por galón.

Este gráfico supone que nuestros datos son una muestra aleatoria de todos los coches de EE.UU., Europa y Japón.

### Ejemplo 

Utilice relplot() y el DataFrame mpg para crear un gráfico de dispersión con "aceleración" en el eje x y "mpg" en el eje y. Varía el estilo y el color de los puntos del gráfico según el país de origen ("origin").

```python
sns.relplot(x="acceleration", y="mpg", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin")

# Show plot
plt.show()
```

```python
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin",
            dashes=False,
            markers=True)

# Show plot
plt.show()
```
