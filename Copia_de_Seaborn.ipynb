# **Relplot**

Gr√°ficos de relaci√≥n entre variables num√©ricas

En este ejercicio, exploraremos el conjunto de datos mpg, que contiene una fila por modelo de coche e incluye informaci√≥n como el a√±o de fabricaci√≥n del coche, el n√∫mero de millas por gal√≥n ("M.P.G.") que alcanza, la potencia de su motor (medida en "caballos") y su pa√≠s de origen.

¬øCu√°l es la relaci√≥n entre la potencia del motor de un coche ("caballos") y su eficiencia de combustible ("mpg")? ¬øY c√≥mo var√≠a esta relaci√≥n seg√∫n el n√∫mero de cilindros ("cylinders") que tenga el coche?

```python
sns.relplot(data=mpg,x="horsepower",y="mpg",size="cylinders")

plt.show()
```

```python
#Utilizando el canal de color (hue) para la variable cylinders

sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders",
            hue="cylinders")

# Show plot
plt.show()
```

## Lineplot

Otro tipo de gr√°fico que nos permite evaluar tendencia.

¬øCu√°l ha sido el comportamiento de mpg a trav√©s del tiempo?

```python
sns.relplot(x="model_year", y="mpg", 
            data=mpg, kind="line")

plt.show()
```

**Pregunta**
¬øCu√°l de las siguientes NO es una interpretaci√≥n correcta de este gr√°fico lineal?

*Posibles respuestas*

El promedio de millas por gal√≥n ha aumentado generalmente con el tiempo.

La distribuci√≥n de millas por gal√≥n es menor en 1973 comparado con 1977.

El intervalo de confianza del 95% para el promedio de millas por gal√≥n en 1970 es aproximadamente 16 - 19.5 millas por gal√≥n.

Este gr√°fico supone que nuestros datos son una muestra aleatoria de todos los coches de EE.UU., Europa y Jap√≥n.

### Ejemplo 

Utilice relplot() y el DataFrame mpg para crear un gr√°fico de dispersi√≥n con "aceleraci√≥n" en el eje x y "mpg" en el eje y. Var√≠a el estilo y el color de los puntos del gr√°fico seg√∫n el pa√≠s de origen ("origin").

```python
sns.relplot(x="acceleration", y="mpg", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin")

# Show plot
plt.show()
```

```python
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin",
            dashes=False,
            markers=True)

# Show plot
plt.show()
```

# **Catplot**

Para visualizar variables categ√≥ricas

```python
sns.catplot(
    data=mpg, x="origin", y="mpg",
    kind="bar"
)
```

```python
sns.catplot(
    data=mpg, x="origin", y="mpg",
    kind="box"
)
```
